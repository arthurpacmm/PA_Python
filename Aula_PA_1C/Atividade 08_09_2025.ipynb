Solicite ao usuário para inserir o nome e a altura de várias pessoas. Preencha um dicionário com essas informações e depois encontre e imprima o nome e a altura da pessoa mais alta e da mais baixa.


```python
alturas = {}

while True:
  nome = input("Digite o nome da pessoa para cadastro ou sair: ").lower()

  if nome == "sair":
    pessoamaisalta = max(alturas, key=alturas.get)
    pessoamaisbaixa = min(alturas, key=alturas.get)
    break
  altura = float(input(f"Digite a altura de {nome}: "))

  alturas[nome]= altura
  print(alturas)

print(f"A pessoa mais alta é {pessoamaisalta} com altura de {alturas[pessoamaisalta]}")
```

Crie uma lista com cinco números inteiros. Use um laço de repetição e uma condicional para imprimir apenas os números que são múltiplos de 3.

```python
numeros = [5, 9, 12, 7, 18]

print("Números múltiplos de 3 na lista:")

for num in numeros:
    if num % 3 == 0:
        print(num)
```

Simule uma situação de alfândega onde o funcionário cadastra itens novos encontrados na mala de uma pessoa que veio de uma viagem internacional. Após o cadastro de todos os itens, o programa imprime uma nota com todos os itens e seus preços, calcula o valor total e, se o valor total ultrapassar 500 dólares, calcula o valor de 60% a ser pago.

```python
produtos = {}

while True:
  itens = input("digite o nome do produto ou digite sair para sair: ")
  if itens == "sair":

    break
  valor = float(input("digite o preço do produto: "))

  produtos[itens] = valor

print(produtos)
total = sum(produtos.values())
print("O valor total é: ",total)
if total >500:
  total = total+(0.6*total)
print("O valor com taxa é: ",total)
```

Você vai criar um programa em Python para analisar um texto e identificar palavras que expressam pensamentos negativos. Para isso, você usará uma lista de palavras negativas fornecida e verificará quais dessas palavras estão presentes no texto fornecido pelo usuário. Exemplo de palavras negtivas, acrescente mais na lista: # Lista de palavras negativas palavras_negativas = ['triste', 'deprimido', 'ansioso', 'desapontado', 'frustrado']

```python
negativas =  ['triste', 'deprimido', 'ansioso', 'desapontado', 'frustrado',]
frase = input("Digite sua frase") .lower()
palavra_negativa = False

for i in negativas:
  if i in frase:
    palavra_negativa = True
    break

if palavra_negativa == True:
  print("Na sua frase tem palavras negativas" , negativas)
else:
  print("Na sua frase não tem palavras negativas")
```
