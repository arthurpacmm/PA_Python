Soma dos números de uma lista
Dada a lista numeros = [3, 7, 2, 8, 10], use um for para calcular a soma de todos os elementos.

```python
# Entrada

numeros = [3, 7, 2, 8, 10]

# Processamento

soma = 0
for numero in numeros:
    soma += numero

# Saída

print("A soma dos elementos é:", soma)
```


Contagem de vogais em uma string
Peça uma palavra ao usuário e use um for para contar quantas vogais (a, e, i, o, u) existem.



```python
# Entrada

palavra = input("Digite uma palavra: ")

vogais = "aeiouAEIOU"
contador = 0

# Processamento

for letra in palavra:
    if letra in vogais:
        contador += 1

# Saída

print("Quantidade de vogais:", contador)
```

Inverter lista usando while
Dada a lista nomes = ["Ana", "João", "Carlos", "Maria"], use um while para inverter a ordem dos elementos (sem usar reverse() ou slicing).

```python
# Entrada

nomes = ["Ana", "João", "Carlos", "Maria"]

invertida = []
i = len(nomes) - 1

# Processamento

while i >= 0:
    invertida.append(nomes[i])
    i -= 1

# Saída

print("Lista invertida:", invertida)
```

Maior número em uma lista
Escreva um programa que percorra uma lista de inteiros e encontre o maior valor.

```python
# Entrada

numeros = [5, 12, 7, 30, 25, 8]

maior = numeros[0]

# Processamento

for numero in numeros:
    if numero > maior:
        maior = numero

# Saída

print("O maior número é:", maior)
```

Remover elementos repetidos de uma lista
Usando um for, percorra a lista valores = [1, 2, 2, 3, 4, 4, 5] e crie uma nova lista apenas com elementos únicos.

```python
# Entrada

valores = [1, 2, 2, 3, 4, 4, 5]

unicos = []

# Processamento

for v in valores:
    if v not in unicos:
        unicos.append(v)

# Saída

print("Lista sem repetidos:", unicos)
```


Verificar se palavra é palíndromo
Peça uma palavra ao usuário e use while ou for para verificar se ela é igual de frente para trás (ex.: "arara").

```python
# Entrada

palavra = input("Digite uma palavra: ")

eh_palindromo = True
tamanho = len(palavra)

# Processamento

for i in range(tamanho // 2):  # só precisa comparar até a metade
    if palavra[i] != palavra[tamanho - 1 - i]:
        eh_palindromo = False
        break

# Saída

if eh_palindromo:
    print("É um palíndromo!")
else:
    print("Não é um palíndromo.")
```


Contagem regressiva com while

Peça ao usuário um número e faça uma contagem regressiva até 0.

```python
# Entrada, Processamento e Saída

n = int(input("Digite um número para a contagem regressiva: "))

while n >= 0:
    print(n)
    n -= 1
```

Filtrar números pares de uma lista

Dada a lista numeros = [10, 21, 32, 43, 54], use um for para criar uma nova lista contendo apenas os pares.

```python
# Entrada

numeros = [10, 21, 32, 43, 54]

pares = []

# Processamento

for n in numeros:
    if n % 2 == 0:
        pares.append(n)

# Saída

print("Números pares:", pares)
```

Contador de palavras em uma frase
Peça ao usuário uma frase, divida em palavras (split()) e use um dicionário para contar quantas vezes cada palavra aparece.

```python
frase = input("Digite uma frase: ")

palavras = frase.split()
contador = {}

for palavra in palavras:
    if palavra in contador:
        contador[palavra] += 1
    else:
        contador[palavra] = 1

print("Contagem de palavras:")
for palavra, qtd in contador.items():
    print(palavra, ":", qtd)
```
