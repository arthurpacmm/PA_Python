Peça ao usuário para digitar seu nome completo e, em seguida, exiba:
1. O nome com todas as letras maiúsculas.
2. O nome com todas as letras minúsculas.
3. Quantas letras tem ao todo (sem considerar espaços).
4. Quantas letras tem o primeiro nome.

Dica. .upper(), .lower(), .replace() e .split()

Conte quantas vogais existem na string
Dica: sum, for, if
Entrada: texto = exercício
Saída; 5

```python
nomec = input("Digite seu nome completo: ")

# 1. Nome em maiúsculas
print("Nome em maiúsculas:", nomec.upper())

# 2. Nome em minúsculas
print("Nome em minúsculas:", nomec.lower())

# 3. Quantidade de letras (sem contar espaços)
letras = len(nomec.replace(" ", ""))
print("Quantidade de letras:", letras)

# 4. Quantidade de letras do primeiro nome
primeiro_nome = nomec.split()[0]
print("Letras no primeiro nome:", len(primeiro_nome))

# 5. Contar vogais
vogais = "aeiouAEIOU"
qvogais = sum(1 for letra in nomec if letra in vogais)
print("Quantidade de vogais:", qvogais)
```
