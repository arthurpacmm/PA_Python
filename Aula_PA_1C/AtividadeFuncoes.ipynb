Questão 8

```python
def calculadora(a, b, operacao):
    if operacao == 'soma':
        return a + b
    elif operacao == 'subtracao':
        return a - b
    elif operacao == 'multiplicacao':
        return a * b
    elif operacao == 'divisao':
        if b == 0:
            return "Erro: divisão por zero!"
        return a / b
    else:
        return "Operação inválida"
```
